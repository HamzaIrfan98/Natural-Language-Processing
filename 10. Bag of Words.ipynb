```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("spam.csv")
df.head()
```

```python
df.Category.value_counts()
```

```python
df['spam'] = df['Category'].apply(lambda x: 1 if x =='spam' else 0)
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2)
```

```python
X_train[:5]
```

```python
y_train[:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()
```

```python
X_train_cv = v.fit_transform(X_train)
X_train_cv
```

```python
X_train_cv.toarray()
```

```python
v.get_feature_names()[1500]
```

```python
v.vocabulary_
```

```python
X_train_np = X_train_cv.toarray()
X_train_np[0]
```

```python
np.where(X_train_np[0]!=0)
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```

```python
emails = [
    'Hey tabish, can we get together to watch pak v ind match tomorrow?',
    'Upto 20% discount on all products, exclusive offer just for you. Dont miss this reward!'
]

emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
# Model can also be trained using the pre-trained sklearn pipeline


from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```
