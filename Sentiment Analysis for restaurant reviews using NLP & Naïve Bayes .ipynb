# Sentiment Analysis for Restaurent Reviews

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# .tsv --> tab seperated value


dataset = pd.read_csv('Desktop/CAT/Data Science with Python/Datasets/Restaurant_Reviews.tsv', delimiter='\t', quoting=3)
```

```python
dataset
```

```python
#Both will help in Data Cleaning

import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer    # will keep the root word only; 'love'; removes 'loved','loves' etc 
```

```python
corpus = []
for i in range(0,1000):
    review = re.sub('[^a-zA-Z]',' ' , dataset['Review'][i])   #Any thing other than a-z or A-Z will be replaced with a space
    review.lower()
    review = review.split()
    ps = PorterStemmer()
    all_stopwords = stopwords.words('english')
    all_stopwords.remove('not')
    review = [ps.stem(word) for word in review if not word in set(all_stopwords)]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer    #Converting our strings into vectors form, to provide to model
```

```python
cv = CountVectorizer(max_features =1500)
x = cv.fit_transform(corpus).toarray()
y = dataset.iloc[:, -1].values
```

```python
len(x)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state =0)
```

```python
X_train
```

```python
X_train.shape
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, Y_train)
```

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), Y_test.reshape(len(Y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test, y_pred)
print(cm)
accuracy = accuracy_score(Y_test, y_pred)
print(accuracy)
```
